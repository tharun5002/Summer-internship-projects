Create a Truck Risk Score.

    Formula:

Risk=0.35(overload ratio)+0.25(avg delay)+0.25(fuel inefficiency)+0.15(failure probability)

```python
overload_ratio = total_weight_delivered / capacity

min_delivery_time = np.min(average_delivery_time_per_truck)
max_delivery_time = np.max(average_delivery_time_per_truck)
avg_delay_normalized = (average_delivery_time_per_truck - min_delivery_time) / (max_delivery_time - min_delivery_time)

fuel_inefficiency = 1 / (efficiency_score + 1e-6)
min_fuel_ineff = np.min(fuel_inefficiency)
max_fuel_ineff = np.max(fuel_inefficiency)
fuel_inefficiency_normalized = (fuel_inefficiency - min_fuel_ineff) / (max_fuel_ineff - min_fuel_ineff)

np.random.seed(42)
failure_probability = np.random.rand(500)

truck_risk_score = (
    0.35 * overload_ratio +
    0.25 * avg_delay_normalized +
    0.25 * fuel_inefficiency_normalized +
    0.15 * failure_probability
)

print("first 10 truck risk scores:", truck_risk_score[:10])
print("max risk rcore:", np.max(truck_risk_score))
print("min risk score:", np.min(truck_risk_score))
print("average risk score:", np.mean(truck_risk_score))
```

Implement mini NumPy functions manually:

Create your own version of:

    mean
    std
    normalization

Without using NumPy built-ins.

```python
def custom_mean(arr):
    flat_arr = arr.flatten() if hasattr(arr, 'flatten') else np.array(arr).flatten()
    total_sum = 0
    for x in flat_arr:
        total_sum += x
    return total_sum / len(flat_arr)

def custom_std(arr):
    flat_arr = arr.flatten() if hasattr(arr, 'flatten') else np.array(arr).flatten()
    mean_val = custom_mean(flat_arr)
    sum_sq_diff = 0
    for x in flat_arr:
        sum_sq_diff += (x - mean_val)**2
    variance = sum_sq_diff / len(flat_arr)
    return variance**0.5

def custom_normalize(arr):
    flat_arr = arr.flatten() if hasattr(arr, 'flatten') else np.array(arr).flatten()
    min_val = flat_arr[0]
    max_val = flat_arr[0]
    for x in flat_arr:
        if x < min_val:
            min_val = x
        if x > max_val:
            max_val = x

    if max_val == min_val:
        return np.zeros_like(arr, dtype=float)

    normalized_arr = (arr - min_val) / (max_val - min_val)
    return normalized_arr


print(f"original distance mean (NumPy): {np.mean(distance):.2f}")
print(f"custom mean of distance: {custom_mean(distance):.2f}")

print(f"original distance std (NumPy): {np.std(distance):.2f}")
print(f"custom std of distance: {custom_std(distance):.2f}")

sample_distance = distance[0, :5]
normalized_sample = custom_normalize(sample_distance)
print(f"sample distance for normalization: {sample_distance}")
print(f"normalized sample distance: {normalized_sample}")
```
